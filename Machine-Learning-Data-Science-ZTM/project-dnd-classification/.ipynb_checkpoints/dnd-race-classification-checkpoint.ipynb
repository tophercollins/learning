## 1. Problem Definition

Given a baseline of character ability scores and general features, can we predict the race of the character?

## 2. Data

> Original data found on Kaggle provided by Andrew Abeles

> https://www.kaggle.com/datasets/andrewabeles/dnd-stats

## 3. Evaluation
> We'll be satisfied if we can match or improve on the Andrew Abeles's previous KNN tuned model of 51.8% accuracy.

## 4. Features

Our dataset has 10000 samples, each with 9 input features and 1 target feature.

#### Input Features:

* height (inches)
* weight (lbs)
* speed (ft.)
* strength
* dexterity
* constitution
* wisdom
* intelligence
* charisma

#### Target Variable:
* race (dragonborn, dwarf, elf, gnome, half-elf, half-orc, halfling, human, tiefling)

## 5. Modelling

```python
# Split data into X and y

X = df.drop('race', axis=1)
y = df['race']
```

```python
# Spit into train and test split
np.random.seed(42)

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
# Create a models dictionary

models = {'SVC' : SVC(kernel='linear'),
         'KNN' : KNeighborsClassifier(),
         'Random Forest' : RandomForestClassifier(),
         'Ada Boost' : AdaBoostClassifier(),
         'Logistic Regression' : LogisticRegression(max_iter=2500)}

# Create a function to fit and score models

def fit_and_score(models, X_train, X_test, y_train, y_test):
    """
    Fits and evaluates machine learning models.
    models: a dict of different Scikit-Learn models
    """
    # Set random seed
    np.random.seed(42)
    # Make a dictionary to keep model scores
    models_scores = {}
    # Loop through the models
    for name, model in models.items():
        # Fit the model to the data
        model.fit(X_train, y_train)
        # Evaluate the model and apped its score the model_scores
        models_scores[name] = model.score(X_test, y_test)
    return models_scores
```

```python
model_scores = fit_and_score(models, X_train, X_test, y_train, y_test)

model_scores
```

Weakest performing are KNN and AdaBoost, so we'll drop them.

Let's fine tune the other 3.

### Intial Tuning Random Forest

```python
# Let's tune Random Forest as it was the lowest of the 3 remaining to see if we can improve.

train_scores = []
test_scores = []

# Create a list of different values for n_estimators
n_estimators = range(50, 800, 50)

# Set up Random Forest
rf = RandomForestClassifier()

# Loop through different n_estimators
for i in n_estimators:
    np.random.seed(42)
    
    rf.set_params(n_estimators=i)
    
    # Fit the algorithm
    rf.fit(X_train, y_train)
    
    # Update the training scores list
    train_scores.append(rf.score(X_train, y_train))
    
    # Update the test scores list
    test_scores.append(rf.score(X_test, y_test))

test_scores
```

```python
plt.plot(n_estimators, test_scores, label="Test score")
plt.xlabel("Number of Estimators")
plt.ylabel("Model score")
plt.legend()

print(f"Maximum Random Forest score on the test data: {max(test_scores)*100:.2f}%")
```

Peak of 67.25% for 250 estimators.

In line with our best baseline models for SVC and Logistic Regression.

### Intial Tuning SVC

```python
# Let's tune SVC to see if we can improve.

train_scores = []
test_scores = []

# Create a list of different values for c
c = [0.01,0.1,1,5]

# Set up SVC
svc = SVC()

# Loop through different c number
for i in c:
    np.random.seed(42)
    
    svc.set_params(kernel='linear', C=i)
    
    # Fit the algorithm
    svc.fit(X_train, y_train)
    
    # Update the training scores list
    train_scores.append(svc.score(X_train, y_train))
    
    # Update the test scores list
    test_scores.append(svc.score(X_test, y_test))

test_scores
```

```python
plt.plot(c, test_scores, label="Test score")
plt.xlabel("C Number")
plt.ylabel("Model score")
plt.legend()

print(f"Maximum SVC score on the test data: {max(test_scores)*100:.2f}%")
```

Peak of 67.15% for both C number 1 & 5.

### Intial Tuning Logistic Regression

```python
# Let's tune Logistic Regression to see if we can improve.

train_scores = []
test_scores = []

# Create a list of different values for c
c = [0.01,0.1,1,5,10]

# Set up Logistic Regression
logreg = LogisticRegression(max_iter=7500)

# Loop through different c number
for i in c:
    np.random.seed(42)

    logreg.set_params(C=i)

    # Fit the algorithm
    logreg.fit(X_train, y_train)

    # Update the training scores list
    train_scores.append(logreg.score(X_train, y_train))

    # Update the test scores list
    test_scores.append(logreg.score(X_test, y_test))

test_scores
```

```python
plt.plot(c, test_scores, label="Test score")
plt.xlabel("C Number")
plt.ylabel("Model score")
plt.legend()

print(f"Maximum Logistic Regression score on the test data: {max(test_scores)*100:.2f}%")
```

Peak of 67.15% with C Number of 0.1, matching the SVC best number. Lets also try with different max_iter tuning.

```python
# Let's tune Logistic Regression to see if we can improve.

train_scores = []
test_scores = []

# Create a list of different values for number of iterations
iterations = [100,500,1000,5000,10000,20000]

# Set up Logistic Regression
logreg = LogisticRegression(C=0.1)

# Loop through different c number
for i in iterations:
    np.random.seed(42)

    logreg.set_params(max_iter=i)

    # Fit the algorithm
    logreg.fit(X_train, y_train)

    # Update the training scores list
    train_scores.append(logreg.score(X_train, y_train))

    # Update the test scores list
    test_scores.append(logreg.score(X_test, y_test))

test_scores
```

```python
plt.plot(iterations, test_scores, label="Test score")
plt.xlabel("Max Number of Iterations")
plt.ylabel("Model score")
plt.legend()

print(f"Maximum Logistic Regression score on the test data: {max(test_scores)*100:.2f}%")
```

Peak of 67.20% with 5000 iterations.

### Initial Tuning Conclusions

#### Model Scores

* Random Forest Classifier - 67.25%
* Logistic Regression - 67.20%
* SVC - 67.15%

Our models all performed nearly identically. Logistic Regresson and SVC has more similarities on application, so we'll dismiss SVC and continue with Randomized and Grid Searches for Random Forest and Logistic Regression.

### Random Forest Randomized Search CV

```python
np.random.seed(42)

param_dist = {
    'n_estimators': [200, 225, 250, 275, 300],
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2'],
    'bootstrap': [True, False]
}

# Create a Random Forest Classifier instance
rf = RandomForestClassifier()

# Create a RandomizedSearchCV instance
random_search = RandomizedSearchCV(rf, param_distributions=param_dist, n_iter=100, cv=5, verbose=2, n_jobs=-1)

# Fit the RandomizedSearchCV on your data
random_search.fit(X_train, y_train);
```

### New Random Forest Classifier with Best Parameters

```python
grid_search.best_params_

clf = RandomForestClassifier(bootstrap=True,
                            criterion='entropy',
                            max_depth=25,
                            max_features='log2',
                            min_samples_leaf=1,
                            min_samples_split=8,
                            n_estimators=225)

# Define the scorers for different metrics with the appropriate average parameter
scoring = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score, average='weighted'),  # Choose appropriate average
    'recall': make_scorer(recall_score, average='weighted'),  # Choose appropriate average
    'f1': make_scorer(f1_score, average='weighted')  # Choose appropriate average
}

# Perform cross-validation for each metric
cv_results = {}
for metric in scoring:
    cv_results[metric] = cross_val_score(clf, X, y, scoring=scoring[metric], cv=5)  # You can adjust the cv value

# Calculate the mean scores for each metric
cv_mean_scores = {metric: np.mean(scores) for metric, scores in cv_results.items()}

# Visualize the cross-validated metrics
cv_metrics = pd.DataFrame(cv_mean_scores, index=[0])
print(cv_metrics)
```

```python
# Define the metrics and colors
metrics = cv_metrics.columns
colors = ['blue', 'green', 'orange', 'red']

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 6))  # Adjust the figure size as needed

# Iterate through each metric and plot a bar with a specific color
bars = []
for i, metric in enumerate(metrics):
    bar = ax.bar(metric, cv_metrics[metric], color=colors[i], width=0.5)  # Adjust the width as needed
    bars.append(bar[0])

# Customize the plot
ax.set_xlabel('Metrics')
ax.set_ylabel('Mean Score')
ax.set_title('Random Forest Classifier Cross-Validated Metrics')
ax.set_xticks(metrics)
ax.set_xticklabels(metrics)
ax.set_ylim(0, 1)  # Set y-axis limits

# Add legend
ax.legend(bars, metrics)

# Adjust the layout to make space for the legend
plt.tight_layout(rect=[0, 0.1, 1, 1])

# Show the plot
plt.show()
```
