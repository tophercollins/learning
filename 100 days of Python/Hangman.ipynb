# Hangman Game!!

```python
import random

word_list = ['orange','monkey','dynamite','politics','library']

def game(word):
    
    playing = True
    life = 6
    display_word = []
    wrong_letters = []
    for letter in word:
        display_word += ['_']
    
    while playing:
        print('\nWord:')
        print(display_word)
        print('\nWrong Letters:')
        print(wrong_letters)
        print(f'Current Lives: {life}')
    
        user_letter = input('Choose your next letter:').lower()
        if user_letter in word:
            for index, letter in enumerate(word):
                if letter == user_letter:
                    display_word[index] = letter.upper()
            if '_' not in display_word:
                print('\nWord:')
                print(display_word)
                print('\nYou Win! You found the missing word!')
                return True
        else:
            life -= 1
            wrong_letters += [user_letter.upper()]
            
            if life <= 0:
                print('\nWord:')
                print(display_word)
                print('\nWrong Letters:')
                print(wrong_letters)
                print(f'Current Lives: {life}')
                print('\nYou Lost! Your hangman died!')
                return True


infinite_loop = True

while infinite_loop:
    game(random.choice(word_list)) 
    input('\n\nWould you like to play again?')
                   
    
```
