* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement

# Logistic Regression

Now it's time to do a train test split, and train our model!

You'll have the freedom here to choose columns that you want to train on!

** Split the data into training set and testing set using train_test_split**

```python
X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male','Timestamp']]
y = ad_data['Clicked on Ad']
```

```python
def adjust_timestamp(df):
    date = [x.split(' ')[0] for x in df['Timestamp']]
    time = [x.split(' ')[1] for x in df['Timestamp']]
    df['Year'] = [int(x.split('-')[0]) for x in date]
    df['Month']= [int(x.split('-')[1]) for x in date]
    df['Day'] = [int(x.split('-')[2]) for x in date]
    df['Hour'] = [int(x.split(':')[0]) for x in time]
    df['Minute'] = [int(x.split(':')[1]) for x in time]
    df.drop('Timestamp',axis=1,inplace=True)
```

```python
adjust_timestamp(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

** Train and fit a logistic regression model on the training set.**

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
pd.DataFrame(lr.coef_[0],index=X.columns,columns=['Coefficents'])
```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
y_preds = lr.predict(X_test)
```

** Create a classification report for the model.**

```python
lr.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(y_test, y_preds))
print(confusion_matrix(y_test, y_preds))
```
